## Chapter 2 — Sediment Contamination Assessment and Screening for Sites with Minimal Contamination Effects on Taxa Composition

Chapter 2 assigns each of the 310 sites a comparable contamination measure built from 16 chemical variables, then uses it to identify a **least-polluted reference subset** for downstream community analysis. **Two** frameworks deliver this: one constructs the contamination stressors and scores, the other validates the selected reference sites.

### Contamination assessment from chemical data.

Reduces the 16 chemicals into rotated PCA stressors and combines them into per-site contamination scores with different views to compare in ranking and selecting the least-polluted candidates.

<!-- explanation -->
The Loadings define how the stressors are linearly constructed from the original chemical variables.
Drop off the $PC_3$ that mainly carries natural metals, the remaining $PCs (16 \times 4)$ left multiplies the chemical matrix $(311 \times 16)$ to bring the most original variation into a $\mathbb{R}^{311 \times 4}$ of four orthogonal stressors, which helps partition the contamination variation in a quantitative cleaner way.

Two views are adopted to process the four stressor values into per-site contamination scores.

View one - additive contamination effect. It assumes stressor effects on community are additive, then four stressor values of each site are summed up in raw scale.
It produces **Sum Relative Contamination Score (SumRel)**.

View two - maximum contamination effect. It assumes stressor effects on community are dominated by the maximum one, then only the maximum stressor value of each site is taken as the contamination score.
It produces **Max Relative Contamination Score (MaxRel)**.

### Framework 2 — Validation of least contamination effects on taxa composition at reference sites

Tests whether the selected least-stressed subset behaves as a defensible baseline by repeatedly sampling random site subsets of equal size. For each subset, RDA regresses taxa composition on stressors and contamination scores, recording adjusted R2R^2
R2 and permutation p-values. The reference subset is validated if its chemical explainability sits near the lower tail of the random distribution. (59 words)
