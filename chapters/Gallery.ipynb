## Chapter 2 — Sediment Contamination Assessment and Screening for Sites with Minimal Contamination Effects on Taxa Composition

Chapter 2 assigns each of the 310 sites a comparable contamination measure built from 16 chemical variables, then uses it to identify a least-polluted reference subset for downstream community analysis. Two frameworks deliver this: one constructs the contamination stressors and scores, the other validates the selected reference sites.

### Contamination assessment from chemical data

Reduces the 16 chemicals into rotated PCA stressors and combines them into per-site contamination scores with different views to compare in ranking and selecting the least-polluted candidates.

Starting with the chemical data matrix $\mathbf{M}_{(311 \times 16)}$, each chemical variable is log-transformed to reduce the influence of extreme raw records. The transformed matrix is then converted into a correlation matrix $\mathbf{cor}_{(16 \times 16)}$, a similarity matrix among variables that implicitly standardizes their raw scales.

PCA decomposes $\mathbf{cor}_{(16 \times 16)}$ to derive new component axes as linear combinations of the original chemical variables. These axes are orthogonal and summarize the main inter-chemical correlation structure. Each chemical variable is represented by its loadings spreading across these component axes.

Further quartimax rotation post-processes the retained PCA loading matrix by multiplying it with an optimized orthogonal rotation matrix. This reorients the retained component axes within the same component space and produces a simpler and more chemical-focused loading pattern.

$$
\mathbf{M}_{(311 \times 16)} \xrightarrow{log_{10}(x + 1) \text{ } \& \text{ cor}} \mathbf{cor}_{(16 \times 16)} \xrightarrow{\text{PCA, }\text{ loading rotation } \& \text{ selection}} \mathbf{PC} \text{ } \mathbf{loadings}_{(16 \times 5)}
$$

> **Same row-wise similarity measures but different column-wise referred objects**
> 
> I prefer to interpret the row-wise records in both $\mathbf{cor}_{(16 \times 16)}$ and $\mathbf{PC} \text{ } \mathbf{loadings}_{(16 \times 16)}$ as chemical similarity measures to the column-wise object. In $\mathbf{cor}_{(16 \times 16)}$, a chemical column has its similarity to chemicals saved in its row positions, so does a loading column in $\mathbf{PC} \text{ } \mathbf{loadings}_{(16 \times 16)}$. Differently, the row-wise records (similarities) are highly condensed in first several column loadings in $\mathbf{PC} \text{ } \mathbf{loadings}_{(16 \times 16)}$, and they are orthogonal to each other.

The Loadings define how the stressors are linearly constructed from the original chemical variables.
Drop off the $PC_2$ that mainly carries natural metals, the remaining $PCs (16 \times 4)$ left multiplies the chemical matrix $(311 \times 16)$ that are normalized by log-transformation and z-score standardization, it brings the most original variation into a $\mathbb{R}^{311 \times 4}$ of four orthogonal stressors, which helps partition the contamination variation in a quantitative cleaner way.

$$
\mathbf{M}_{(311 \times 16)}^{'} \mathbf{PCs}_{(16 \times 4)} = \mathbf{S}_{(311 \times 4)} = [S_1, S_2, S_3, S_4]_{(311 \times 4)}
\tag{1}
$$

On top of the four stressors values, two views are adopted to process the four stressor values into per-site contamination scores.

1. **Additive contamination effect**. It assumes stressor effects on community are additive, then four stressor values of each site are summed up in raw scale.
It produces Sum Relative Contamination Score (**SumRel**).

$$
\text{SumRel} = \mathbf{S}_{(311 \times 4)} \mathbf{1}_{(4 \times 1)} = \mathbf{C}_{(311 \times 1)} = [c_1, c_2, ..., c_{311}]^T_{(311 \times 1)}

\tag{2}
$$

2. **Maximum contamination effect**. It assumes stressor effects on community are dominated by the maximum one, then only the maximum stressor value of each site is taken as the contamination score.
It produces Max Relative Contamination Score (MaxRel).

$$
\text{MaxRel} = \max_{\text{column wise}}(\mathbf{S}_{(311 \times 4)}) = \mathbf{C}_{(311 \times 1)} = [c_1, c_2, ..., c_{311}]^T_{(311 \times 1)}

\tag{3}
$$

### Validate least contamination effects on taxa composition of the selected reference sites

With the SumRel contamination scores $\mathbf{C}_{(311 \times 1)}$, it supports to rank and select the sites with lowest contamination scores as the least-polluted sites where the taxa compositions are assumed to be least affected by compound contamination.

In previous work (Zhang, 2008) it was made without validatoin that the 20% sites (62) with lowest contamination scores were selected as least-polluted sites, where the taxa composition was assumed in reference condition with least contamination effects.

In this framework, still the 20% (62) lowest-scored sites are selected as the community reference subset but their defensibility are validated by repeatedly resampling random site subsets of equal size. For each subset, RDA (a multivariate regression method) regresses taxa matrix on retained stressor matrix and records $\bar R^2$ as indicator of the stressor constrained variation in taxa matrix. 

The chosen 62 sites should have a significantly low $\bar R^2$ in a random distribution of $\bar R^2$ from the resampled subsets, which interprets few stressor constrained variation in their taxa matrix and defends them as least contamination affected sites in terms of taxa composition.


A one-sided hypothesis test is conducted, and the derived $p$-value is evaluated under $H_0$ through this process:

<div style="width: 80%; max-width: 700px; margin: 0 auto;">

| $H_i$ | Statement |
|:---:|:---|
| $H_0$ | The 62 lowest-scored sites do not show weaker stressor-constrained variation than a random 62-site subset. |
| $H_A$ | The 62 lowest-scored sites show weaker stressor-constrained variation than a random 62-site subset. |

</div>

<div style="width: 80%; max-width: 700px; margin: 0 auto;">

![Empirical distribution — adjusted R²](../codespace/figures/ch2_fw2_empirical_distribution_adj_R2.png)

*Figure 1: Empirical distribution of $\bar R^2$ from 1000 randomly resampled 62-site subsets. The vertical black line marks the empirically extreme observed $\bar R^2$ for the 62 lowest-scored sites; its associated $p$-value is $\le 0.05$, so $H_0$ is rejected.*

In addition to the empirical $\bar R^2$ distribution from the resampled subsets, a permutation test is also applied to each subset and to the 62 lowest-scored sites.

For each 62-site subset with observed pairs of $(\mathbf{T}_i, \mathbf{S}_i)$, the null hypothesis assumes that there are no contamination effects on taxa composition and is evaluated by shuffling the stressor values across the observed group assignments. Each subset is permuted 1000 times to generate a null distribution of $\bar R^2$ under the assumption of no contamination effects, and the observed $\bar R^2$ of the subset is then compared with this null distribution to derive a $p$-value for the hypothesis test.

For a specific subset, the hypothesis test underlying this $p$-value is:

<div style="width: 80%; max-width: 700px; margin: 0 auto;">

| $H_i$ | Statement |
|:---:|:---|
| $H_0$ | There are no stressor-constrained effects on taxa composition in the observed subset. |
| $H_A$ | There are stressor-constrained effects on taxa composition in the observed subset. |

</div>

As $m_{\text{ref}}$ increases, more contaminated sites are included in the lowest-$m_{\text{ref}}$ reference subset, and higher $\bar R^2$ values are observed, as reflected by the rightward movement of the circles along the x-axis.

The probabilistic extremeness of the observed $\bar R^2$ can still be significant across these cases, even though the absolute values differ.

Beyond assessing the defensibility of the lowest-scored subset within the random-subset distribution for the same $m_{\text{ref}}$, this comparison also raises the question of which $m_{\text{ref}}$ is optimal for defining the reference subset, or what extent of stressor-constrained variation in taxa composition is acceptable as a trade-off against reference-subset size.

## Chapter 3 — Environmental Standardization by Reference Community Types

Chapter 3 builds an environmental standardization framework for zoobenthic community assessment. It first identifies reference community types from least-polluted sites via unconstrained clustering on their taxa features, then filters those reference sites by taxa–environment concordance to retain high-confidence training cases. Finally, it trains an environmental classifier to assign all sites their expected community type under low-contamination conditions, creating the baseline needed to interpret contamination-driven deviations in community composition.

### Reference Community Types from Least-Polluted Sites

Starting from the acknowledgment that the least-polluted sites have community compositions least affected by contamination, they are expected to reflect the natural variation in community composition driven by environmental gradients.
Partitioning these naturally varying communities into discrete community types is a gentle (safe) and interpretation-friendly way to separate their variation into distinct groups.
This approach allows normal fluctuation in community composition within groups while preserving natural separation between groups.

**Unconstrained hierarchical clustering** is applied to the taxa matrix of the least-polluted reference sites to identify clusters as reference community types without using any environmental descriptors.
Ward's clustering defines the linkage function by minimizing the total within-cluster variance at each merging step, so it expects meaningful Euclidean distances between sites derived from the taxa matrix.
To retain ecological realism in a Euclidean-defined distance space, Hellinger transformation is applied to the default proportional-octave transformed taxa matrix so that
the Euclidean distance between Hellinger-transformed sites can approximate their proportional compositional dissimilarity regardless of the original taxa abundance scales.

Figure <span style="color: blue;">1</span> shows the Ward dendrogram for the 62 least-polluted sites at $k=2$.
The linkage distance at the first split, which bifurcates the 62 sites into two clusters, is around 3.0 and much larger than the second linkage distance of around 1.8, indicating a strong separation between the first two clusters.

#### `pvclust` with AU values for cluster robustness evaluation

The **AU values** shown in the legend are branch-level support values from the `pvclust` analysis. In this workflow, `pvclust` first constructs a main dendrogram from the original, non-bootstrap taxa matrix using the same clustering settings as the initial Ward clustering. The internal branches of this main dendrogram define the candidate clusters to be evaluated. Bootstrap and multiscale-bootstrap resampling are then applied to the descriptor features, and the data are reclustered repeatedly to evaluate whether the same branch-defined site clusters are recovered across the resampled dendrograms.

Because the transposed taxa matrix is supplied to `pvclust`, the columns represent the 62 sites being clustered, while the rows represent the 16 taxa descriptors being resampled. Therefore, the AU values evaluate the robustness of the site clusters under resampling of taxa features, rather than under resampling of sites.

The AU value is an approximately unbiased support value obtained from multiscale bootstrap resampling, where different bootstrap sample sizes are used to reduce the bias of ordinary bootstrap probability. Another BP value is the ordinary bootstrap probability, measuring how often the same branch-defined cluster is recovered under ordinary bootstrap resampling. In `pvclust`, larger AU values indicate stronger support for the corresponding cluster. A common guideline is that clusters with AU values near or above 0.95 are strongly supported.

When the AU value decreases, it may indicate that the cluster is supported by fewer consistent taxa features. However, a community type with weak statistical support can still be ecologically meaningful. The AU value should be treated as evidence of clustering robustness, not as the only criterion for ecological relevance.

The computation was implemented using the `pvclust` package in R, called from Python through the `rpy2` interface to keep the workflow consistent with the rest of the Python-based analysis.

This main dendrogram serves as the reference tree for evaluating cluster recovery across bootstrap and multiscale-bootstrap resampling. The two-cluster Ward solution shown in Figure 1 was compared with the branch structure of this main `pvclust` dendrogram.

In this comparison, `Match = exact` and `Jaccard = 1` indicate that each finalized Ward cluster had exactly the same site membership as a corresponding branch in the main `pvclust` dendrogram. Therefore, the AU and BP values reported for those matched branches can be interpreted as bootstrap support values for the two finalized Ward clusters.

The AU values were relatively high for both clusters ($C_1$: 0.927; $C_2$: 0.940), suggesting moderate-to-strong support under multiscale bootstrap resampling of the 16 taxa descriptors, although they are slightly below the common 0.95 guideline for strong cluster support. The ordinary BP values were much lower ($C_1$: 0.055; $C_2$: 0.050), indicating that the same branches were rarely recovered under ordinary bootstrap resampling. Because AU is adjusted by multiscale bootstrap to reduce the bias of ordinary BP, AU is used as the primary support measure here.

#### Node-wise ANOVA of taxa at splitting nodes in the dendrogram

In the Ward dendrogram, each internal node represents a split between two daughter branches, denoted here as the **left** and **right** groups. Although the linkage distance is computed from site-level dissimilarities, the separation between two branches can be interpreted by checking which taxa differ most between the two groups. To examine this, node-wise ANOVA was applied to the Hellinger-transformed taxa values at selected splitting nodes. For each taxon $x$, the mean values $\bar{x}_{left}$ and $\bar{x}_{right}$ were compared between the left and right branches, and the resulting $F$-statistic and $p$-value indicate the strength of the between-branch difference. The $p$ FDR column reports the false-discovery-rate adjusted $p$-value, used to control for multiple taxa tested at the same node.

Table <span style="color: blue;">2</span> summarizes the significant taxa at three important splitting nodes around the finalized $k=2$ solution. Split 1 separates all 62 sites into the finalized groups of 46 and 16 sites. This split is mainly characterized by higher **Dreissena**, **Hydrozoa**, **Turbellaria**, **Hydropsychidae**, and **Nematoda** in the right branch, and higher **Chironomidae** and **Oligochaeta** in the left branch. Therefore, Split 1 represents the main compositional contrast between the two finalized reference community types.

Split 2 further divides the 16-site right branch into groups of 2 and 14 sites. This secondary split is mainly associated with **Dreissena** and **Nematoda**, suggesting additional within-type variation inside the Dreissena-associated community. Split 3 further divides the 46-site left branch into groups of 22 and 24 sites, mainly driven by **Oligochaeta**, **Chironomidae**, **Dreissena**, **Amphipoda**, and **Caenis**. These secondary splits indicate that natural site-level variation remains within each finalized community type.

> Branch size is also important when interpreting node-wise splits. Very small daughter branches, such as groups with fewer than about 10 sites, may provide unstable benchmarks for a reference community type. With limited sites, the average taxa profile and its association with environmental descriptors are strongly influenced by individual-site variation and hard to train a robust classifier.

#### Ecological interpretation of the taxa profiles of $C_1$ and $C_2$

Figure <span style="color: blue;">2</span> visualizes the taxa profiles of the two clusters based on the One-way ANOVA results. The bar plots are the averages of hellinger-transformed abundance of the 16 taxa with $1$-SE error bars, showing contrast compositions of **soft-sediment benthic** taxa and **substrate-associated** taxa between the two clusters.

$C_1$ (blue) is a soft-sediment benthic community type with high abundant Oligochaeta and Chironomidae, which are common in large-river depositional habitats with low flow and fine sediments.
$C_2$ (orange) is a substrate-associated community type with high abundant Dreissena and relatively higher Hydrozoa, Turbellaria than in $C_1$, suggesting a higher-flow or deeper habitat conditions than $C_1$.

The significance markers highlight taxa with clearer **between-cluster** differences. Each cluster is characterized by an unique compositional signal: $C_1$ is mainly distinguished by higher relative abundance of soft-sediment benthic taxa, whereas $C_2$ is mainly distinguished by the stronger contribution of substrate-associated taxa. 
The error bars show that **site-level variation** driven by fine-scale natural conditions still remains within each community type. Therefore, the two clusters should be interpreted as reference community types with distinct dominant taxa profiles, while still allowing natural variation among individual site communities.

### Association with Environmental Profiles - Subsets of Taxa-Environment Concordance

Previous work were around defining community types as finalized clusters from the least-polluted sites, regardless the environmental profiles of them. It is an ecologically factual mechanism that countless environmental factors of a habitat mainly and commonly shape and present the community composition that we observed at a site.
Confined by the availability and observability of environmental descriptors, there always have limited environmental descriptors to make up the environmental space in which we can settle the sites to interprete their naturally driven community variation.

Stick with the fact that true environmental condition dominates the community composition with fine-scale variation on individual taxa-level, then the derivative community types should stay consistent with corresponding environmental profiles. However, an nonneglectable problem arising from the trade-off between choosing the complexity depth of community types and which environmental descriptors that make up the environmental space to support that environment-to-taxa interpretation.
