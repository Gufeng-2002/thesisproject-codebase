# Records

Throughout the research project, weekly meetings were held with supervisors and collaborators, including Dr. Jan Ciborowski, Dr. Jabed Tomal, Dr. Yue Zhang, and Jian Zhang. The following sections provide access to the meeting notes, documenting the project's progress since May 2025.

Each note begins with a one-sentence summary for quick browsing. Click a note to view the full details.
Each note is organized into three sections:

- **Discussion**: Topics discussed during the meeting.

- **Feedback**: Comments and guidance provided by the supervisors.

- **Action Items**: Specific plans and follow-up tasks identified after the meeting.

This section also includes special notes, such as the thesis summary submitted to the 2026 Annual Conference of the Statistical Society of Canada (SSC), along with related materials. These can be found under their corresponding titles.
