<div class="hero">
  <h1>Environmental Assessment of Zoobenthic Communities</h1>
  <p class="subtitle">A computational framework for sediment contamination assessment and community-condition indexing</p>
</div>

<div class="overview">
  <p>
    This project develops a reproducible analytical pipeline for assessing the ecological condition
    of zoobenthic (bottom-dwelling invertebrate) communities in relation to sediment contamination.
    The work integrates three types of site-level data&mdash;environmental habitat descriptors,
    sediment chemistry, and taxonomic community composition&mdash;to build a bioassessment framework
    that separates natural environmental variation from contamination-driven impacts.
  </p>
  <p>
    The pipeline processes <strong>310 sites</strong> with complete data across three matrices:
  </p>

  <div class="matrix-table">
    <table>
      <thead>
        <tr><th>Data Matrix</th><th>Symbol</th><th>Dimension</th><th>Units</th></tr>
      </thead>
      <tbody>
        <tr><td>Environmental</td><td><strong>E</strong></td><td>310 &times; 6</td><td>mixed</td></tr>
        <tr><td>Chemical</td><td><strong>M</strong></td><td>310 &times; 16</td><td>mixed</td></tr>
        <tr><td>Taxa</td><td><strong>T</strong></td><td>310 &times; 16</td><td>octave</td></tr>
      </tbody>
    </table>
  </div>
