# Definition Corner

Define the reusable table-joining, regression-fitting, and plotting helpers used to compare cluster-specific ZCI values against contamination scores.

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import warnings
from statsmodels.tools.sm_exceptions import IterationLimitWarning


QUANTILE_LEVELS = [0.8, 0.5, 0.2]
REGRESSION_LINE_COLORS = {
    "OLS": "#111111",
    "Q0.8": "#d95f02",
    "Q0.5": "#1b9e77",
    "Q0.2": "#7570b3",
}


PLANE_QUANTILE_LINE_VISIBILITY = {
    0.8: False,
    0.5: True,
    0.2: False,
}


def _normalize_quantile_model_name(quantile_level: float | str) -> str:
    if isinstance(quantile_level, str):
        return quantile_level if quantile_level.startswith("Q") else f"Q{float(quantile_level):.1f}"
    return f"Q{float(quantile_level):.1f}"


def _resolve_quantile_line_visibility(
    quantile_line_visibility: dict[float | str, bool] | None = None,
) -> dict[str, bool]:
    visibility_map = {
        _normalize_quantile_model_name(quantile_level): bool(is_visible)
        for quantile_level, is_visible in PLANE_QUANTILE_LINE_VISIBILITY.items()
    }
    if quantile_line_visibility is None:
        return visibility_map
    visibility_map.update(
        {
            _normalize_quantile_model_name(quantile_level): bool(is_visible)
            for quantile_level, is_visible in quantile_line_visibility.items()
        }
    )
    return visibility_map


def build_zci_stressor_table(
    zci_table: pd.DataFrame,
    stressor_table: pd.DataFrame,
    score_col: str = CONTAMINATION_SCORE,
    id_col: str = SITE_ID_COLUMN,
) -> pd.DataFrame:
    required_zci_columns = {id_col, "ZCI"}
    missing_zci_columns = required_zci_columns.difference(zci_table.columns)
    if missing_zci_columns:
        raise KeyError(f"The ZCI table is missing required columns: {sorted(missing_zci_columns)}")
    required_stressor_columns = {id_col, score_col}
    missing_stressor_columns = required_stressor_columns.difference(stressor_table.columns)
    if missing_stressor_columns:
        raise KeyError(f"The stressor table is missing required columns: {sorted(missing_stressor_columns)}")

    zci_subset = zci_table[[id_col, "ZCI"]].copy()
    zci_subset[id_col] = zci_subset[id_col].astype(str)
    score_subset = stressor_table[[id_col, score_col]].copy()
    score_subset[id_col] = score_subset[id_col].astype(str)
    if score_subset[id_col].duplicated().any():
        duplicated_ids = score_subset.loc[score_subset[id_col].duplicated(), id_col].unique().tolist()[:5]
        raise ValueError(f"Duplicate stressor records found for site IDs: {duplicated_ids}")

    merged_table = zci_subset.merge(score_subset, on=id_col, how="left", validate="one_to_one")
    missing_scores = merged_table.loc[merged_table[score_col].isna(), id_col].tolist()
    if missing_scores:
        raise ValueError(f"Missing {score_col} values for site IDs: {missing_scores[:10]}")

    merged_table[score_col] = pd.to_numeric(merged_table[score_col], errors="coerce")
    merged_table["ZCI"] = pd.to_numeric(merged_table["ZCI"], errors="coerce")
    if merged_table[[score_col, "ZCI"]].isna().any().any():
        raise ValueError(f"The merged {score_col}-ZCI table contains missing or non-numeric values.")
    return merged_table[[id_col, score_col, "ZCI"]]


def _piecewise_design_matrix(x_values: np.ndarray, knot: float) -> np.ndarray:
    return np.column_stack([
        np.ones_like(x_values, dtype=float),
        x_values,
        np.maximum(0.0, x_values - knot),
    ])


def _quantile_loss(residuals: np.ndarray, tau: float) -> float:
    return float(np.sum(np.where(residuals >= 0.0, tau * residuals, (tau - 1.0) * residuals)))


def _resolve_breakpoint_candidate_knots(
    x_values: np.ndarray,
    candidate_count: int,
    min_side_count: int,
    breakpoint_search_interval: tuple[float, float] | None = None,
) -> np.ndarray:
    if breakpoint_search_interval is None:
        candidate_knots = np.unique(np.quantile(x_values, np.linspace(0.15, 0.85, candidate_count)))
    else:
        interval_low, interval_high = map(float, breakpoint_search_interval)
        if not interval_low < interval_high:
            raise ValueError("breakpoint_search_interval must be an increasing (low, high) tuple.")
        candidate_knots = np.linspace(interval_low, interval_high, candidate_count)

    candidate_knots = np.array([
        float(knot)
        for knot in candidate_knots
        if (x_values <= knot).sum() >= min_side_count and (x_values > knot).sum() >= min_side_count
    ], dtype=float)

    if candidate_knots.size == 0 and breakpoint_search_interval is not None:
        interval_low, interval_high = map(float, breakpoint_search_interval)
        bounded_x = x_values[(x_values >= interval_low) & (x_values <= interval_high)]
        if bounded_x.size > 0:
            candidate_knots = np.unique(np.quantile(bounded_x, np.linspace(0.1, 0.9, candidate_count)))
            candidate_knots = np.array([
                float(knot)
                for knot in candidate_knots
                if (x_values <= knot).sum() >= min_side_count and (x_values > knot).sum() >= min_side_count
            ], dtype=float)

    if candidate_knots.size == 0:
        candidate_knots = np.array([float(np.median(x_values))])
    return candidate_knots


def fit_ols_regression(x_values: np.ndarray, y_values: np.ndarray) -> dict:
    design_matrix = sm.add_constant(x_values)
    model = sm.OLS(y_values, design_matrix).fit()
    slope = float(model.params[1])
    intercept = float(model.params[0])
    return {
        "model": "OLS",
        "tau": np.nan,
        "knot": np.nan,
        "intercept": intercept,
        "slope_left": slope,
        "slope_right": slope,
        "r_squared": float(model.rsquared),
        "predictor": lambda grid, intercept=intercept, slope=slope: intercept + slope * grid,
    }


def fit_piecewise_quantile_regression(
    x_values: np.ndarray,
    y_values: np.ndarray,
    tau: float,
    candidate_count: int = 31,
    breakpoint_search_interval: tuple[float, float] | None = None,
) -> dict:
    min_side_count = max(5, int(np.ceil(0.1 * x_values.size)))
    candidate_knots = _resolve_breakpoint_candidate_knots(
        x_values,
        candidate_count=candidate_count,
        min_side_count=min_side_count,
        breakpoint_search_interval=breakpoint_search_interval,
    )

    best_fit = None
    best_loss = np.inf
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", IterationLimitWarning)
        for knot in candidate_knots:
            design_matrix = _piecewise_design_matrix(x_values, knot)
            try:
                model = sm.QuantReg(y_values, design_matrix).fit(q=tau, max_iter=10000, p_tol=1e-7)
            except Exception:
                continue
            predicted_values = design_matrix @ model.params
            loss = _quantile_loss(y_values - predicted_values, tau)
            if loss < best_loss:
                best_fit = (float(knot), model)
                best_loss = loss

    if best_fit is None:
        raise ValueError(f"Could not fit a piecewise quantile regression for tau={tau}.")

    knot, model = best_fit
    intercept = float(model.params[0])
    slope_left = float(model.params[1])
    slope_change = float(model.params[2])
    slope_right = slope_left + slope_change
    baseline_prediction = float(np.quantile(y_values, tau))
    baseline_loss = _quantile_loss(y_values - baseline_prediction, tau)
    pseudo_r_squared = 1.0 - best_loss / baseline_loss if baseline_loss > 0.0 else np.nan

    return {
        "model": f"Q{tau:.1f}",
        "tau": float(tau),
        "knot": float(knot),
        "intercept": intercept,
        "slope_left": slope_left,
        "slope_right": slope_right,
        "r_squared": float(pseudo_r_squared),
        "predictor": lambda grid, intercept=intercept, slope_left=slope_left, slope_change=slope_change, knot=knot: (
            intercept + slope_left * grid + slope_change * np.maximum(0.0, grid - knot)
        ),
    }


def fit_cluster_regressions(
    zci_stressor_table: pd.DataFrame,
    score_col: str = CONTAMINATION_SCORE,
    quantile_levels: list[float] | None = None,
    breakpoint_search_interval: tuple[float, float] | None = None,
) -> list[dict]:
    quantile_levels = quantile_levels or QUANTILE_LEVELS
    working_table = zci_stressor_table[[score_col, "ZCI"]].copy()
    working_table[score_col] = pd.to_numeric(working_table[score_col], errors="coerce")
    working_table["ZCI"] = pd.to_numeric(working_table["ZCI"], errors="coerce")
    working_table = working_table.dropna().sort_values(score_col)
    if working_table.shape[0] < 6:
        raise ValueError("At least six matched observations are required for the regression comparison.")
    if np.isclose(working_table[score_col].var(ddof=0), 0.0):
        raise ValueError(f"The {score_col} values must vary to fit regressions.")

    x_values = working_table[score_col].to_numpy(dtype=float)
    y_values = working_table["ZCI"].to_numpy(dtype=float)
    fitted_models = [fit_ols_regression(x_values, y_values)]
    fitted_models.extend(
        fit_piecewise_quantile_regression(
            x_values,
            y_values,
            tau=tau,
            breakpoint_search_interval=breakpoint_search_interval,
        )
        for tau in quantile_levels
    )
    return fitted_models


def summarize_regression_models(
    fitted_models: list[dict],
    cluster_label: str,
) -> pd.DataFrame:
    return pd.DataFrame([
        {
            "Cluster": cluster_label,
            "Model": fitted_model["model"],
            "Tau": fitted_model["tau"],
            "Breakpoint SumRel": fitted_model["knot"],
            "Left slope": fitted_model["slope_left"],
            "Right slope": fitted_model["slope_right"],
            "R2": fitted_model["r_squared"],
        }
        for fitted_model in fitted_models
    ])


def render_cluster_regression_plane(
    zci_stressor_table: pd.DataFrame,
    fitted_models: list[dict],
    cluster_label: str,
    quantile_line_visibility: dict[float | str, bool] | None = None,
    score_col: str = CONTAMINATION_SCORE,
    dpi: int = 300,
) -> tuple[plt.Figure, plt.Axes]:
    working_table = zci_stressor_table[[score_col, "ZCI"]].dropna().sort_values(score_col)
    x_values = working_table[score_col].to_numpy(dtype=float)
    y_values = working_table["ZCI"].to_numpy(dtype=float)
    x_padding = 0.04 * (x_values.max() - x_values.min()) if x_values.max() > x_values.min() else 1.0
    line_grid = np.linspace(x_values.min(), x_values.max(), 300)

    figure, axis = plt.subplots(figsize=(8.0, 5.6), dpi=dpi)
    axis.scatter(
        x_values,
        y_values,
        s=34,
        color="#4c78a8",
        alpha=0.76,
        edgecolor="white",
        linewidth=0.35,
        label="Sites",
    )

    resolved_quantile_line_visibility = _resolve_quantile_line_visibility(quantile_line_visibility)

    for fitted_model in fitted_models:
        model_name = fitted_model["model"]
        if model_name != "OLS" and not resolved_quantile_line_visibility.get(model_name, True):
            continue
        line_style = "--" if model_name == "OLS" else "-"
        line_width = 2.0 if model_name == "OLS" else 1.8
        line_color = REGRESSION_LINE_COLORS.get(model_name, "#555555")
        r_squared = fitted_model["r_squared"]
        label = f"{model_name} R²={r_squared:.3f}"
        if model_name != "OLS":
            label = f"{model_name} piecewise R²={r_squared:.3f}"
        axis.plot(
            line_grid,
            fitted_model["predictor"](line_grid),
            linestyle=line_style,
            linewidth=line_width,
            color=line_color,
            label=label,
        )

    axis.set_xlim(x_values.min() - x_padding, x_values.max() + x_padding)
    axis.set_xlabel(f"Contamination score ({score_col})", fontsize=12, fontweight="bold")
    axis.set_ylabel("ZCI", fontsize=12, fontweight="bold")
    axis.set_title(f"{cluster_label} ZCI-Contamination Regression Plane", fontsize=14, fontweight="bold")
    axis.grid(True, linestyle="--", linewidth=0.8, alpha=0.35)
    axis.set_axisbelow(True)
    axis.legend(frameon=True, fontsize=9)
    figure.tight_layout()
    return figure, axis
```

```python
def bootstrap_piecewise_quantile_regression(
    zci_stressor_table: pd.DataFrame,
    quantile_levels: list[float],
    bootstrap_replicates: int,
    random_seed: int,
    score_col: str = CONTAMINATION_SCORE,
    candidate_count: int = 15,
    breakpoint_search_intervals: dict[float, tuple[float, float]] | None = None,
) -> pd.DataFrame:
    working_table = zci_stressor_table[[score_col, "ZCI"]].copy()
    working_table[score_col] = pd.to_numeric(working_table[score_col], errors="coerce")
    working_table["ZCI"] = pd.to_numeric(working_table["ZCI"], errors="coerce")
    working_table = working_table.dropna().sort_values(score_col)
    if working_table.shape[0] < 6:
        raise ValueError("At least six matched observations are required for bootstrap breakpoint estimation.")

    x_values = working_table[score_col].to_numpy(dtype=float)
    y_values = working_table["ZCI"].to_numpy(dtype=float)
    rng = np.random.default_rng(random_seed)
    bootstrap_rows = []

    for tau in quantile_levels:
        successful_replicates = 0
        attempts = 0
        max_attempts = max(bootstrap_replicates * 5, bootstrap_replicates + 25)
        while successful_replicates < bootstrap_replicates and attempts < max_attempts:
            attempts += 1
            sampled_index = rng.integers(0, x_values.size, size=x_values.size)
            sample_x = x_values[sampled_index]
            sample_y = y_values[sampled_index]
            if np.unique(sample_x).size < 6 or np.isclose(np.var(sample_x), 0.0):
                continue
            try:
                fitted_model = fit_piecewise_quantile_regression(
                    sample_x,
                    sample_y,
                    tau=tau,
                    candidate_count=candidate_count,
                    search_interval=_resolve_tau_breakpoint_search_interval(breakpoint_search_intervals, tau),
                )
            except Exception:
                continue

            slope_change = fitted_model["slope_right"] - fitted_model["slope_left"]
            bootstrap_rows.append(
                {
                    "Tau": float(tau),
                    "Replicate": successful_replicates + 1,
                    "Breakpoint SumRel": float(fitted_model["knot"]),
                    "Intercept": float(fitted_model["intercept"]),
                    "Left slope": float(fitted_model["slope_left"]),
                    "Slope change": float(slope_change),
                    "Right slope": float(fitted_model["slope_right"]),
                    "R2": float(fitted_model["r_squared"]),
                }
            )
            successful_replicates += 1

        if successful_replicates < bootstrap_replicates:
            raise ValueError(
                f"Only {successful_replicates} successful bootstrap fits were obtained for tau={tau}."
            )

    return pd.DataFrame(bootstrap_rows)


def summarize_bootstrap_piecewise_quantile_models(
    bootstrap_results: pd.DataFrame,
    cluster_label: str,
    confidence_level: float,
) -> pd.DataFrame:
    if not 0.0 < confidence_level < 1.0:
        raise ValueError("confidence_level must be strictly between 0 and 1.")
    lower_tail_probability = (1.0 - confidence_level) / 2.0
    upper_tail_probability = 1.0 - lower_tail_probability

    summary_rows = []
    for tau, tau_table in bootstrap_results.groupby("Tau", sort=False):
        summary_rows.append(
            {
                "Cluster": cluster_label,
                "Model": f"Q{tau:.1f}",
                "Tau": float(tau),
                "Confidence level": float(confidence_level),
                "Successful bootstraps": int(tau_table.shape[0]),
                "Median breakpoint": float(tau_table["Breakpoint SumRel"].median()),
                "Breakpoint CI low": float(tau_table["Breakpoint SumRel"].quantile(lower_tail_probability)),
                "Breakpoint CI high": float(tau_table["Breakpoint SumRel"].quantile(upper_tail_probability)),
                "Median intercept": float(tau_table["Intercept"].median()),
                "Median left slope": float(tau_table["Left slope"].median()),
                "Median slope change": float(tau_table["Slope change"].median()),
                "Median right slope": float(tau_table["Right slope"].median()),
                "Median R2": float(tau_table["R2"].median()),
            }
        )
    return pd.DataFrame(summary_rows).sort_values("Tau", ascending=False).reset_index(drop=True)


def build_cluster_pole_plane_points(
    stressor_table: pd.DataFrame,
    cluster_label: str,
    score_col: str = CONTAMINATION_SCORE,
    id_col: str = SITE_ID_COLUMN,
) -> pd.DataFrame:
    required_columns = {id_col, score_col}
    missing_columns = required_columns.difference(stressor_table.columns)
    if missing_columns:
        raise KeyError(f"The stressor table is missing required columns: {sorted(missing_columns)}")

    pole_zci_map = {
        f"{cluster_label} ref pole": 1.0,
        f"{cluster_label} deg pole": 0.0,
    }
    score_subset = stressor_table[[id_col, score_col]].copy()
    score_subset[id_col] = score_subset[id_col].astype(str)
    pole_points = score_subset.loc[score_subset[id_col].isin(pole_zci_map)].copy()
    missing_poles = [pole_id for pole_id in pole_zci_map if pole_id not in pole_points[id_col].tolist()]
    if missing_poles:
        raise KeyError(f"Missing pole rows in the stressor table: {missing_poles}")

    pole_points = pole_points.set_index(id_col).loc[list(pole_zci_map)].reset_index()
    pole_points[score_col] = pd.to_numeric(pole_points[score_col], errors="coerce")
    if pole_points[score_col].isna().any():
        raise ValueError(f"Pole contamination scores in {score_col} must be numeric.")
    pole_points["ZCI"] = pole_points[id_col].map(pole_zci_map)
    return pole_points[[id_col, score_col, "ZCI"]]


def _estimate_gaussian_score_density(sample_values: np.ndarray, grid: np.ndarray) -> np.ndarray:
    sample_values = np.asarray(sample_values, dtype=float)
    grid = np.asarray(grid, dtype=float)
    if sample_values.size == 0:
        raise ValueError("At least one sample is required for density estimation.")

    if sample_values.size == 1:
        sample_scale = max(float(np.ptp(grid)) / 60.0, 1.0)
    else:
        sample_std = float(np.std(sample_values, ddof=1))
        sample_iqr = float(np.subtract(*np.percentile(sample_values, [75.0, 25.0])))
        robust_scale = sample_iqr / 1.349 if sample_iqr > 0.0 else 0.0
        sample_scale = max(sample_std, robust_scale, max(float(np.ptp(sample_values)) / 10.0, 1.0e-6))

    bandwidth = 1.06 * sample_scale * max(sample_values.size, 2) ** (-1.0 / 5.0)
    bandwidth = max(float(bandwidth), max(float(np.ptp(grid)) / 80.0, 0.35))
    standardized_distances = (grid[:, None] - sample_values[None, :]) / bandwidth
    kernel_values = np.exp(-0.5 * standardized_distances**2) / (np.sqrt(2.0 * np.pi) * bandwidth)
    return kernel_values.mean(axis=1)


def render_bootstrap_piecewise_quantile_plane(
    zci_stressor_table: pd.DataFrame,
    bootstrap_summary: pd.DataFrame,
    cluster_label: str,
    confidence_level: float,
    pole_points: pd.DataFrame | None = None,
    conditional_density_specs: list[dict] | None = None,
    quantile_line_visibility: dict[float | str, bool] | None = None,
    score_col: str = CONTAMINATION_SCORE,
    dpi: int = 300,
) -> tuple[plt.Figure, plt.Axes]:
    working_table = zci_stressor_table[[score_col, "ZCI"]].dropna().sort_values(score_col)
    x_values = working_table[score_col].to_numpy(dtype=float)
    y_values = working_table["ZCI"].to_numpy(dtype=float)
    x_padding = 0.04 * (x_values.max() - x_values.min()) if x_values.max() > x_values.min() else 1.0
    line_grid = np.linspace(x_values.min(), x_values.max(), 300)

    figure, axis = plt.subplots(figsize=(8.0, 5.6), dpi=dpi)
    axis.scatter(
        x_values,
        y_values,
        s=34,
        color="#4c78a8",
        alpha=0.72,
        edgecolor="white",
        linewidth=0.35,
        zorder=2,
    )

    if pole_points is not None and not pole_points.empty:
        pole_colors = {
            f"{cluster_label} ref pole": "#2ca25f",
            f"{cluster_label} deg pole": "#de2d26",
        }
        for pole_id, pole_color in pole_colors.items():
            pole_row = pole_points.loc[pole_points[SITE_ID_COLUMN] == pole_id]
            if pole_row.empty:
                continue
            axis.scatter(
                pole_row[score_col].to_numpy(dtype=float),
                pole_row["ZCI"].to_numpy(dtype=float),
                marker="*",
                s=220,
                color=pole_color,
                edgecolor="#111111",
                linewidth=0.8,
                zorder=5,
            )

    resolved_quantile_line_visibility = _resolve_quantile_line_visibility(quantile_line_visibility)

    for summary_row in bootstrap_summary.sort_values("Tau", ascending=False).to_dict("records"):
        model_name = summary_row["Model"]
        if not resolved_quantile_line_visibility.get(model_name, True):
            continue
        line_color = REGRESSION_LINE_COLORS.get(model_name, "#555555")
        median_breakpoint = summary_row["Median breakpoint"]
        median_intercept = summary_row["Median intercept"]
        median_left_slope = summary_row["Median left slope"]
        median_slope_change = summary_row["Median slope change"]
        predicted_values = (
            median_intercept
            + median_left_slope * line_grid
            + median_slope_change * np.maximum(0.0, line_grid - median_breakpoint)
        )

        axis.axvspan(
            summary_row["Breakpoint CI low"],
            summary_row["Breakpoint CI high"],
            color="#bdbdbd",
            alpha=0.22,
            zorder=0,
        )
        axis.plot(
            line_grid,
            predicted_values,
            color=line_color,
            linewidth=2.0,
            zorder=3,
        )

    if conditional_density_specs:
        y_span = max(float(np.ptp(y_values)), 1.0)
        max_density_height = 0.16 * y_span
        for density_spec in conditional_density_specs:
            baseline_zci = float(density_spec["baseline_zci"])
            zci_low = float(density_spec["zci_low"])
            zci_high = density_spec.get("zci_high")
            if zci_high is None:
                density_mask = working_table["ZCI"] >= zci_low
            else:
                density_mask = (working_table["ZCI"] >= zci_low) & (working_table["ZCI"] <= float(zci_high))

            density_samples = working_table.loc[density_mask, score_col].to_numpy(dtype=float)
            if density_samples.size == 0:
                raise ValueError(
                    f"No contamination scores are available for the conditional density band anchored at ZCI={baseline_zci}."
                )

            density_values = _estimate_gaussian_score_density(density_samples, line_grid)
            peak_density = float(np.max(density_values))
            if not np.isfinite(peak_density) or np.isclose(peak_density, 0.0):
                continue

            density_curve = baseline_zci + max_density_height * density_values / peak_density
            curve_color = density_spec.get("line_color", "#4d4d4d")
            fill_alpha = float(density_spec.get("fill_alpha", 0.12))
            support_mask = density_values >= 0.03 * peak_density

            axis.fill_between(
                line_grid,
                baseline_zci,
                density_curve,
                color=curve_color,
                alpha=fill_alpha,
                linewidth=0.0,
                zorder=1.5,
            )
            if support_mask.any():
                support_grid = line_grid[support_mask]
                axis.hlines(
                    baseline_zci,
                    float(support_grid[0]),
                    float(support_grid[-1]),
                    color=curve_color,
                    linewidth=0.8,
                    alpha=0.7,
                    zorder=2.5,
                )
            axis.plot(
                line_grid,
                density_curve,
                color=curve_color,
                linewidth=1.7,
                zorder=4.5,
            )

    axis.set_xlim(x_values.min() - x_padding, x_values.max() + x_padding)
    axis.set_xlabel(f"Contamination score ({score_col})", fontsize=12, fontweight="bold")
    axis.set_ylabel("ZCI", fontsize=12, fontweight="bold")
    axis.set_title(f"{cluster_label} Bootstrap Piecewise Quantile Regression", fontsize=14, fontweight="bold")
    axis.grid(True, linestyle="--", linewidth=0.8, alpha=0.35)
    axis.set_axisbelow(True)
    figure.tight_layout()
    return figure, axis
```

```python
def infer_breakpoint_search_interval(
    zci_stressor_table: pd.DataFrame,
    score_col: str = CONTAMINATION_SCORE,
    lower_quantile: float = 0.15,
    upper_quantile: float = 0.85,
) -> tuple[float, float]:
    score_values = pd.to_numeric(zci_stressor_table[score_col], errors="coerce").dropna()
    if score_values.empty:
        raise ValueError(f"The {score_col} column must contain at least one numeric value.")
    lower_bound = float(score_values.quantile(lower_quantile))
    upper_bound = float(score_values.quantile(upper_quantile))
    if not np.isfinite(lower_bound) or not np.isfinite(upper_bound) or lower_bound >= upper_bound:
        raise ValueError("The inferred breakpoint search interval is invalid.")
    return lower_bound, upper_bound


def _normalize_breakpoint_search_interval(
    breakpoint_search_interval,
) -> tuple[float, float] | None:
    if breakpoint_search_interval is None:
        return None
    interval_low, interval_high = map(float, breakpoint_search_interval)
    if not interval_low < interval_high:
        raise ValueError("breakpoint_search_interval must be an increasing (low, high) tuple.")
    return interval_low, interval_high


def _resolve_tau_breakpoint_search_interval(
    breakpoint_search_intervals: dict[float, tuple[float, float]] | None,
    tau: float,
) -> tuple[float, float] | None:
    if breakpoint_search_intervals is None:
        return None
    normalized_intervals = {
        float(key): _normalize_breakpoint_search_interval(value)
        for key, value in breakpoint_search_intervals.items()
    }
    tau_key = float(tau)
    if tau_key not in normalized_intervals:
        raise KeyError(f"No breakpoint search interval was provided for tau={tau_key}.")
    return normalized_intervals[tau_key]


def _build_candidate_knots(
    x_values: np.ndarray,
    min_side_count: int,
    candidate_count: int,
    search_interval: tuple[float, float] | None = None,
) -> np.ndarray:
    if search_interval is None:
        lower_bound = float(np.quantile(x_values, 0.15))
        upper_bound = float(np.quantile(x_values, 0.85))
    else:
        lower_bound, upper_bound = _normalize_breakpoint_search_interval(search_interval)

    lower_bound = max(lower_bound, float(np.min(x_values)))
    upper_bound = min(upper_bound, float(np.max(x_values)))
    if not np.isfinite(lower_bound) or not np.isfinite(upper_bound) or lower_bound >= upper_bound:
        raise ValueError("The breakpoint search interval must define a non-empty range inside the observed contamination scores.")

    candidate_knots = np.unique(np.linspace(lower_bound, upper_bound, candidate_count))
    candidate_knots = np.array([
        knot
        for knot in candidate_knots
        if (x_values <= knot).sum() >= min_side_count and (x_values > knot).sum() >= min_side_count
    ], dtype=float)
    if candidate_knots.size == 0:
        candidate_knots = np.array([float(np.median(x_values))])
    return candidate_knots


def fit_piecewise_quantile_regression(
    x_values: np.ndarray,
    y_values: np.ndarray,
    tau: float,
    candidate_count: int = 31,
    search_interval: tuple[float, float] | None = None,
) -> dict:
    min_side_count = max(5, int(np.ceil(0.1 * x_values.size)))
    candidate_knots = _build_candidate_knots(
        x_values,
        min_side_count=min_side_count,
        candidate_count=candidate_count,
        search_interval=search_interval,
    )

    best_fit = None
    best_loss = np.inf
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", IterationLimitWarning)
        for knot in candidate_knots:
            design_matrix = _piecewise_design_matrix(x_values, knot)
            try:
                model = sm.QuantReg(y_values, design_matrix).fit(q=tau, max_iter=10000, p_tol=1e-7)
            except Exception:
                continue
            predicted_values = design_matrix @ model.params
            loss = _quantile_loss(y_values - predicted_values, tau)
            if loss < best_loss:
                best_fit = (float(knot), model)
                best_loss = loss

    if best_fit is None:
        raise ValueError(f"Could not fit a piecewise quantile regression for tau={tau}.")

    knot, model = best_fit
    intercept = float(model.params[0])
    slope_left = float(model.params[1])
    slope_change = float(model.params[2])
    slope_right = slope_left + slope_change
    baseline_prediction = float(np.quantile(y_values, tau))
    baseline_loss = _quantile_loss(y_values - baseline_prediction, tau)
    pseudo_r_squared = 1.0 - best_loss / baseline_loss if baseline_loss > 0.0 else np.nan

    return {
        "model": f"Q{tau:.1f}",
        "tau": float(tau),
        "knot": float(knot),
        "intercept": intercept,
        "slope_left": slope_left,
        "slope_right": slope_right,
        "r_squared": float(pseudo_r_squared),
        "predictor": lambda grid, intercept=intercept, slope_left=slope_left, slope_change=slope_change, knot=knot: (
            intercept + slope_left * grid + slope_change * np.maximum(0.0, grid - knot)
        ),
    }


def fit_cluster_regressions(
    zci_stressor_table: pd.DataFrame,
    score_col: str = CONTAMINATION_SCORE,
    quantile_levels: list[float] | None = None,
    breakpoint_search_intervals: dict[float, tuple[float, float]] | None = None,
) -> list[dict]:
    quantile_levels = quantile_levels or QUANTILE_LEVELS
    working_table = zci_stressor_table[[score_col, "ZCI"]].copy()
    working_table[score_col] = pd.to_numeric(working_table[score_col], errors="coerce")
    working_table["ZCI"] = pd.to_numeric(working_table["ZCI"], errors="coerce")
    working_table = working_table.dropna().sort_values(score_col)
    if working_table.shape[0] < 6:
        raise ValueError("At least six matched observations are required for the regression comparison.")
    if np.isclose(working_table[score_col].var(ddof=0), 0.0):
        raise ValueError(f"The {score_col} values must vary to fit regressions.")

    x_values = working_table[score_col].to_numpy(dtype=float)
    y_values = working_table["ZCI"].to_numpy(dtype=float)
    fitted_models = [fit_ols_regression(x_values, y_values)]
    fitted_models.extend(
        fit_piecewise_quantile_regression(
            x_values,
            y_values,
            tau=tau,
            search_interval=_resolve_tau_breakpoint_search_interval(breakpoint_search_intervals, tau),
        )
        for tau in quantile_levels
    )
    return fitted_models


def bootstrap_piecewise_quantile_regression(
    zci_stressor_table: pd.DataFrame,
    quantile_levels: list[float],
    bootstrap_replicates: int,
    random_seed: int,
    score_col: str = CONTAMINATION_SCORE,
    candidate_count: int = 15,
    breakpoint_search_intervals: dict[float, tuple[float, float]] | None = None,
) -> pd.DataFrame:
    working_table = zci_stressor_table[[score_col, "ZCI"]].copy()
    working_table[score_col] = pd.to_numeric(working_table[score_col], errors="coerce")
    working_table["ZCI"] = pd.to_numeric(working_table["ZCI"], errors="coerce")
    working_table = working_table.dropna().sort_values(score_col)
    if working_table.shape[0] < 6:
        raise ValueError("At least six matched observations are required for bootstrap breakpoint estimation.")

    x_values = working_table[score_col].to_numpy(dtype=float)
    y_values = working_table["ZCI"].to_numpy(dtype=float)
    rng = np.random.default_rng(random_seed)
    bootstrap_rows = []

    for tau in quantile_levels:
        successful_replicates = 0
        attempts = 0
        max_attempts = max(bootstrap_replicates * 5, bootstrap_replicates + 25)
        while successful_replicates < bootstrap_replicates and attempts < max_attempts:
            attempts += 1
            sampled_index = rng.integers(0, x_values.size, size=x_values.size)
            sample_x = x_values[sampled_index]
            sample_y = y_values[sampled_index]
            if np.unique(sample_x).size < 6 or np.isclose(np.var(sample_x), 0.0):
                continue
            try:
                fitted_model = fit_piecewise_quantile_regression(
                    sample_x,
                    sample_y,
                    tau=tau,
                    candidate_count=candidate_count,
                    search_interval=_resolve_tau_breakpoint_search_interval(breakpoint_search_intervals, tau),
                )
            except Exception:
                continue

            slope_change = fitted_model["slope_right"] - fitted_model["slope_left"]
            bootstrap_rows.append(
                {
                    "Tau": float(tau),
                    "Replicate": successful_replicates + 1,
                    "Breakpoint SumRel": float(fitted_model["knot"]),
                    "Intercept": float(fitted_model["intercept"]),
                    "Left slope": float(fitted_model["slope_left"]),
                    "Slope change": float(slope_change),
                    "Right slope": float(fitted_model["slope_right"]),
                    "R2": float(fitted_model["r_squared"]),
                }
            )
            successful_replicates += 1

        if successful_replicates < bootstrap_replicates:
            raise ValueError(
                f"Only {successful_replicates} successful bootstrap fits were obtained for tau={tau}."
            )

    return pd.DataFrame(bootstrap_rows)
```

```python
BOOTSTRAP_CLUSTER_LABELS = ["C1", "C2"]
if STUDY_CASE == "DR":
    BOOTSTRAP_CLUSTER_LABELS.append("C3")

BOOTSTRAP_CONDITIONAL_DENSITY_SPECS = {
    "C1": [
        {
            "baseline_zci": 2.0,
            "zci_low": 2.0,
            "zci_high": None,
            "line_color": "#4d4d4d",
            "fill_alpha": 0.12,
        },
        {
            "baseline_zci": 0.0,
            "zci_low": -0.25,
            "zci_high": 0.0,
            "line_color": "#737373",
            "fill_alpha": 0.12,
        },
    ],
    "C2": [
        {
            "baseline_zci": 0.75,
            "zci_low": 0.75,
            "zci_high": None,
            "line_color": "#4d4d4d",
            "fill_alpha": 0.12,
        },
        {
            "baseline_zci": -0.5,
            "zci_low": -0.5,
            "zci_high": -0.25,
            "line_color": "#737373",
            "fill_alpha": 0.12,
        },
    ],
}
if STUDY_CASE == "DR":
    BOOTSTRAP_CONDITIONAL_DENSITY_SPECS["C3"] = [
        {
            "baseline_zci": 1.5,
            "zci_low": 1.5,
            "zci_high": None,
            "line_color": "#4d4d4d",
            "fill_alpha": 0.12,
        },
        {
            "baseline_zci": -0.5,
            "zci_low": -0.5,
            "zci_high": -0.25,
            "line_color": "#737373",
            "fill_alpha": 0.12,
        },
    ]

c1_tau08_breakpoint_search_interval = [-20, 20]
c1_tau05_breakpoint_search_interval = [-5, 5]
c1_tau02_breakpoint_search_interval = [-5, 5]

c2_tau08_breakpoint_search_interval = [3, 8]
c2_tau05_breakpoint_search_interval = [-5, 5]
c2_tau02_breakpoint_search_interval = [-15, -5]

c3_tau08_breakpoint_search_interval = [3, 8]
c3_tau05_breakpoint_search_interval = [-5, 5]
c3_tau02_breakpoint_search_interval = [-15, -5]

BOOTSTRAP_REPLICATES = 100
BOOTSTRAP_RANDOM_SEED = 20260531
BOOTSTRAP_QUANTILE_LEVELS = [0.8, 0.5, 0.2]
BOOTSTRAP_CANDIDATE_COUNT = 15
BOOTSTRAP_CONFIDENCE_LEVEL = 0.9

if "cluster_inputs" not in globals():
    cluster_inputs = {
        "C1": {
            "zci_table": C1_ZCIs,
            "stressor_table": C1_S_c,
        },
        "C2": {
            "zci_table": C2_ZCIs,
            "stressor_table": C2_S_c,
        },
    }
    if STUDY_CASE == "DR":
        cluster_inputs["C3"] = {
            "zci_table": C3_ZCIs,
            "stressor_table": C3_S_c,
        }

if "cluster_zci_stressor_tables" not in globals():
    cluster_zci_stressor_tables = {}
    for cluster_label, cluster_input in cluster_inputs.items():
        cluster_zci_stressor_tables[cluster_label] = build_zci_stressor_table(
            zci_table=cluster_input["zci_table"],
            stressor_table=cluster_input["stressor_table"],
            score_col=CONTAMINATION_SCORE,
            id_col=SITE_ID_COLUMN,
        )

CLUSTER_TAU_BREAKPOINT_SEARCH_INTERVALS = {
    "C1": {
        0.8: c1_tau08_breakpoint_search_interval,
        0.5: c1_tau05_breakpoint_search_interval,
        0.2: c1_tau02_breakpoint_search_interval,
    },
    "C2": {
        0.8: c2_tau08_breakpoint_search_interval,
        0.5: c2_tau05_breakpoint_search_interval,
        0.2: c2_tau02_breakpoint_search_interval,
    },
}
if STUDY_CASE == "DR":
    CLUSTER_TAU_BREAKPOINT_SEARCH_INTERVALS["C3"] = {
        0.8: c3_tau08_breakpoint_search_interval,
        0.5: c3_tau05_breakpoint_search_interval,
        0.2: c3_tau02_breakpoint_search_interval,
    }

missing_bootstrap_clusters = [
    cluster_label for cluster_label in BOOTSTRAP_CLUSTER_LABELS
    if cluster_label not in cluster_inputs or cluster_label not in CLUSTER_TAU_BREAKPOINT_SEARCH_INTERVALS
]
if missing_bootstrap_clusters:
    raise KeyError(f"Bootstrap cluster configuration is missing for: {missing_bootstrap_clusters}")

cluster_seed_offsets = {
    cluster_label: cluster_index
    for cluster_index, cluster_label in enumerate(cluster_inputs, start=1)
}
cluster_tau_breakpoint_search_interval_table = pd.DataFrame([
    {
        "Parameter": f"{cluster_label.lower()}_{'tau08' if tau == 0.8 else 'tau05' if tau == 0.5 else 'tau02'}_breakpoint_search_interval",
        "Cluster": cluster_label,
        "Tau": tau,
        "Breakpoint search low": search_interval[0],
        "Breakpoint search high": search_interval[1],
    }
    for cluster_label, tau_intervals in CLUSTER_TAU_BREAKPOINT_SEARCH_INTERVALS.items()
    if cluster_label in BOOTSTRAP_CLUSTER_LABELS
    for tau, search_interval in tau_intervals.items()
])

print("Breakpoint search intervals used for piecewise quantile fitting and bootstrap CI")
print(f"Bootstrap confidence level: {BOOTSTRAP_CONFIDENCE_LEVEL:.0%}")
display(cluster_tau_breakpoint_search_interval_table)

bootstrap_breakpoint_search_intervals = {
    cluster_label: {
        tau: tau_intervals[tau]
        for tau in BOOTSTRAP_QUANTILE_LEVELS
    }
    for cluster_label, tau_intervals in CLUSTER_TAU_BREAKPOINT_SEARCH_INTERVALS.items()
    if cluster_label in BOOTSTRAP_CLUSTER_LABELS
}
cluster_bootstrap_pole_points = {
    cluster_label: build_cluster_pole_plane_points(
        cluster_inputs[cluster_label]["stressor_table"],
        cluster_label=cluster_label,
        score_col=CONTAMINATION_SCORE,
        id_col=SITE_ID_COLUMN,
    )
    for cluster_label in BOOTSTRAP_CLUSTER_LABELS
}

cluster_bootstrap_results = {}
cluster_bootstrap_summaries = []
cluster_bootstrap_figures = {}
cluster_bootstrap_axes = {}

with plt.ioff():
    for cluster_label in BOOTSTRAP_CLUSTER_LABELS:
        zci_stressor_table = cluster_zci_stressor_tables[cluster_label]
        bootstrap_results = bootstrap_piecewise_quantile_regression(
            zci_stressor_table,
            quantile_levels=BOOTSTRAP_QUANTILE_LEVELS,
            bootstrap_replicates=BOOTSTRAP_REPLICATES,
            random_seed=BOOTSTRAP_RANDOM_SEED + cluster_seed_offsets[cluster_label],
            score_col=CONTAMINATION_SCORE,
            candidate_count=BOOTSTRAP_CANDIDATE_COUNT,
            breakpoint_search_intervals=bootstrap_breakpoint_search_intervals[cluster_label],
        )
        bootstrap_summary = summarize_bootstrap_piecewise_quantile_models(
            bootstrap_results,
            cluster_label,
            confidence_level=BOOTSTRAP_CONFIDENCE_LEVEL,
        )
        bootstrap_figure, bootstrap_axis = render_bootstrap_piecewise_quantile_plane(
            zci_stressor_table,
            bootstrap_summary,
            cluster_label=cluster_label,
            confidence_level=BOOTSTRAP_CONFIDENCE_LEVEL,
            pole_points=cluster_bootstrap_pole_points[cluster_label],
            conditional_density_specs=BOOTSTRAP_CONDITIONAL_DENSITY_SPECS.get(cluster_label),
            quantile_line_visibility=PLANE_QUANTILE_LINE_VISIBILITY,
            score_col=CONTAMINATION_SCORE,
        )

        cluster_bootstrap_results[cluster_label] = bootstrap_results
        cluster_bootstrap_summaries.append(bootstrap_summary)
        cluster_bootstrap_figures[cluster_label] = bootstrap_figure
        cluster_bootstrap_axes[cluster_label] = bootstrap_axis

cluster_bootstrap_breakpoint_summary = pd.concat(cluster_bootstrap_summaries, ignore_index=True)

print(
    f"Bootstrap breakpoint summary using B={BOOTSTRAP_REPLICATES} pairs resamples, confidence level {BOOTSTRAP_CONFIDENCE_LEVEL:.0%}, and quantile levels {BOOTSTRAP_QUANTILE_LEVELS}"
)
display(cluster_bootstrap_breakpoint_summary)

for cluster_label in BOOTSTRAP_CLUSTER_LABELS:
    print(f"{cluster_label} bootstrap breakpoint plane; matched {cluster_zci_stressor_tables[cluster_label].shape[0]} sites")
    display(cluster_bootstrap_figures[cluster_label])
    plt.close(cluster_bootstrap_figures[cluster_label])
```
