# From Chapter Goals to Computation Pipelines: A Three-Level Thesis Architecture

The narrative way of this study follows a three-level architecture, which are connected and progressive levels.
A discussion around the chapter $i$ with $k_i$ frameworks is in layout as below:

- **Chapter $i$ Layer**

    '*Do what we do with a chapter.*'

  - **Framework $j$ Layer** $(j = 1, \dots, k_i)$
  
    '*Do what we do with a framework.*'

    - **Computation Process $j$ Layer** $(j = 1, \dots, k_i)$
    
      '*Do what we do with a computation process.*'

Their basic roles and relationships are as below:

**Chapter**:
  What larger research objective is being addressed?

**Framework**:
  What analytical logic is used to address part of that objective?

**Computation Process**:
  What exact inputs, tools, models, and outputs implement that framework?

 
This three-level architecture will exist throughout all discussions in the thesis,
specific structures and designs within them are described in the following:

$$
\text{Chapter } i \xrightarrow{contains} \text{Framework } j \xrightarrow{guides} \text{Computation Process } j 
\quad (j = 1, \dots, k_i)
$$


## Chapter $i$ Overview

A chapter starts with a summary that includes the following 5 components:

**Objectives**
 - What this chapter aims to achieve.

**Reasons**
 - Why this chapter is necessary in the thesis.

**How**
 - The high-level reasoning that implements the chapter's objectives.

**Expected Contribution**
 - What this chapter should tell us after completion.

**List of Frameworks**
 - A list of frameworks included in this chapter and the sub-target of each framework.

## Framework $j$ Overview

**Questions/Motivations**
 - What problem this framework addresses, why it matters, and what concerns must be handled.

**Methods and Steps**
 - The major techniques used, their order, and how they interact.

**Expected Conclusions**
 - What kind of result this framework should produce and how it supports the chapter objective.

## Computation Process $j$ Overview

The **input-process-output** process describes every computation flow in this study,
it consists of three objects: (inputs, process, outputs), they form a computation path of:

$$
\text{Inputs} \rightarrow \text{Process} \rightarrow \text{Outputs}
$$

For example, if five chapters involve computation and each chapter builds on earlier results, the same framework can be reused recursively across all five stages.

```text
# inputs for stage 0 have already been prepared
For each stage i in [0, 1, 2, 3, 4]:

    # run the chapter-specific process
    inputs -> process.i -> outputs

    # discuss the outputs in chapter i
    outputs -> discussion in chapter i

    # prepare inputs for stage i + 1
    inputs <- (fetch.chapter(i+1).metadata, outputs.artifacts)
```
